##### 데이터에 맞게 이상치 데이터 처리 -> IQR 이용

```python
Q1 = df1['카드이용금액계(AMT_CORR)'].quantile(.25)
Q3 = df1['카드이용금액계(AMT_CORR)'].quantile(.75)
IQR = Q3 - Q1
Upper_IQR = Q3 + 2.0*IQR
Lower_IQR = Q1 - 2.0*IQR
print(f'Upper Tukey Fences : {Upper_IQR}, Lower Tukey Fences : {Lower_IQR}')
# df1.loc[df1['카드이용금액계(AMT_CORR)']>=Upper_IQR]
df1 = df1.drop(df1.loc[df1['카드이용금액계(AMT_CORR)']>=Upper_IQR].index, inplace = False)
df1
```

```python
Q1 = df3['카드이용금액계(AMT_CORR)'].quantile(.25)
Q3 = df3['카드이용금액계(AMT_CORR)'].quantile(.75)
IQR = Q3 - Q1
Upper_IQR = Q3 + 2.0*IQR
Lower_IQR = Q1 - 2.0*IQR
print(f'Upper Tukey Fences : {Upper_IQR}, Lower Tukey Fences : {Lower_IQR}')
# df1.loc[df1['카드이용금액계(AMT_CORR)']>=Upper_IQR]
df3 = df3.drop(df3.loc[df3['카드이용금액계(AMT_CORR)']>=Upper_IQR].index, inplace = False)
df3
```

```python
Q1 = df4['카드이용금액계(AMT_CORR)'].quantile(.25)
Q3 = df4['카드이용금액계(AMT_CORR)'].quantile(.75)
IQR = Q3 - Q1
Upper_IQR = Q3 + 2.0*IQR
Lower_IQR = Q1 - 2.0*IQR
print(f'Upper Tukey Fences : {Upper_IQR}, Lower Tukey Fences : {Lower_IQR}')
# df1.loc[df1['카드이용금액계(AMT_CORR)']>=Upper_IQR]
df4 = df4.drop(df4.loc[df4['카드이용금액계(AMT_CORR)']>=Upper_IQR].index, inplace = False)
df4
```

##### 요일별 카드이용금액 평균을 plt.plot()을 이용해서 시각화

#### 3-2 통계적 검정

```python
import scipy.stats as stats
```

##### 연령대와 업종 소분류가 연관이 있는지 교차분석

H0 : 연령대와 업종 소분류는 연관이 없다.
H1 : 연령대와 업종 소분류는 연관이 있다.

```python
df1.head()
```

```python
data_cross = pd.crosstab(df1['연령대별(AGE)'], df1['소분류(CLASS3)'])
data_cross
```

```python
result = stats.chi2_contingency(observed = data_cross)
# print(result,'\n')
print(f'''chi-square 값 : {result[0]}
p-value : {result[1]}
자유도 : {result[2]}''')
```

p-value(0.000) <br 0.05 이므로 귀무가설 기각</br>
 - 따라서 연령대와 업종 소분류는 연관이 있다.

##### 남녀간 건당카드이용금액에 차이가 있는지 평균차이를 검정

H0 : 남녀간 건당카드이용금액에 차이가 없다.
H1 : 남녀간 건당카드이용금액에 차이가 있다.

```python
df1.head()
```

```python
#성별에 따라 DataFrame을 분리
df1_f = df1[df1['성별(GENDER)']=='F']
df1_m = df1[df1['성별(GENDER)']=='M']
x = df1_f['건당카드이용금액(MEAN_CORR)']
y = df1_m['건당카드이용금액(MEAN_CORR)']
```

```python
# 두 집단 간의 평균 비교 -> 남/여간 평균카드이용금액 비교
# T-test를 통하여 진행
t_stat, p_val = stats.ttest_ind(x, y)
print(f't-statictic : {t_stat}, p-value : {p_val}')
```

p-value(0.77) > 0.05이므로 귀무가설을 기각하지 못함.

따라서 남녀간 건당카드이용금액에 차이가 없다고 할 수 있다.

##### 상관관계 검정

H0 : 20대의 카드이용건수와 카드이용금액계는 상관관계가 없다.
H1 : 20대의 카드이용건수와 카드이용금액계는 상관관계가 있다.

```python
df1.head()
```

```python
# 20대
y20 = df1[df1['연령대별(AGE)'] == '20대']
y20
```

```python
x = y20['카드이용건수계(USECT_CORR)']
y = y20['카드이용금액계(AMT_CORR)']

plt.scatter(x,y)
plt.show()

c_stat, p_val = stats.pearsonr(x,y)
print(f'상관 계수 : {c_stat}, p-value : {p_val}')
```

p-value(0.000) < 0.05 이므로 귀무가설을 기각할 수 있음.

따라서 20대의 카드이용건수와 카드이용금액계는 상관관계가 있다.

##### 인과관계 검정

H0 : 20대의 카드이용건수와 카드이용금액계는 인과관계가 없다.
H1 : 20대의 카드이용건수와 카드이용금액계는 인과관계가 있다.

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

X = x.values.reshape(-1, 1)

lr.fit(X, y)

print('lr.coef_ : ', lr.coef_)
print('lr.intercept_ : ', lr.intercept_)
```

```python
y2 = lr.coef_[0]*X +lr.intercept_
plt.scatter(x, y)
plt.plot(x, y2, color='red')
plt.title('y = {}*x + {}'.format(lr.coef_[0], lr.intercept_))
plt.show()
```

```python
# 회귀분석 검정
import statsmodels.api as sm
results = sm.OLS(y, sm.add_constant(x)).fit()
results.summary()
```

p-value(0.000) < 0.05 이므로 귀무가설을 기각할 수 있음.

따라서 20대의 카드이용건수와 카드이용금액계는 인과관계가 있다고 할 수 있다.
